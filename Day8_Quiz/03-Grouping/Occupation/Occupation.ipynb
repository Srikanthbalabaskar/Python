### Step 1. Import the necessary libraries

### Step 3. Assign it to a variable called users.

```python
users=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep='|')
users
```

```python
users.dtypes
```

### Step 4. Discover what is the mean age per occupation

```python
meanage=users['age'].mean() #Mean of Ages
meanAgeOcc=users.groupby(['occupation']).agg({'age':np.mean})
meanAgeOcc
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
malerat=users.groupby(['occupation','gender']).agg({'gender':np.size})
malerat
#malerat1=users.groupby(['occupation'],sort='False').
#malerat1
#malerat1=users.groupby(['occupation']).sum().sort_values('gender',ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
#Minimum per occupation
occmin=users.groupby(['occupation']).min()
occmin
```

```python
occmax=users.groupby(['occupation']).max()
occmax
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
occupgen=users.groupby(['occupation','gender']).agg({'age':np.mean})
occupgen
```
