# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Visualization/Titanic_Desaster/train.csv). 

### Step 3. Assign it to a variable titanic 

```python
titanic=pd.read_csv('train.csv')
```

```python
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId',inplace=True)
```

```python
titanic.tail()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
titanic['Sex'].value_counts()
```

```python
plt.pie
```

```python
plt.figure(figsize=(8,6))
plt.pie(titanic['Sex'].value_counts(),labels=['Male','Female'],autopct='%1.1f%%')
plt.title('Male Vs Female Proportion')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Fare',y='Age',hue='Sex',data=titanic)
plt.title('Fare Payed Vs Age based on Gender')
plt.show()
```

### Step 7. How many people survived?

```python
titanic[titanic['Survived']==1].count()['Survived']
```

```python
plt.figure(figsize=(8,6))
plt.pie(titanic['Survived'].value_counts(),labels=['Survived','Not Survived'],autopct='%1.1f%%')
plt.title('Survived Vs Not Survived')
plt.show()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.figure(figsize=(8,6))
plt.title('Histogram of Fare Played')
sns.distplot(a=titanic['Fare'])
plt.show()
```

```python
#Another way to do it
plt.figure(figsize=(8,6))
plt.title('Histogram of Fare Played')
plt.hist(x=titanic['Fare'])
plt.xlabel('Fare')
plt.show()
```
