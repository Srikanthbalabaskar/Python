### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 3. Assign it to a variable called tips

```python
tips=pd.read_csv('tips.csv')
```

```python
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips=tips.drop('Unnamed: 0',axis=1)
```

```python
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(6,4))
sns.distplot(a=tips['total_bill'],kde=True,color='y')
plt.title('Histogram of Total_bill')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
plt.title('Total Bill vs Tip')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],size=tips['tip'])
plt.title('Total Bill vs Tip')
plt.show()
```

```python
#adding hue makes no difference
plt.figure(figsize=(6,4))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],size=tips['tip'],hue=tips['tip'])
plt.title('Total Bill vs Tip')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(6,4))
sns.lineplot(x=tips['day'],y=tips['total_bill'])
plt.title('Days Vs Total Bill')
plt.xlabel('Days')
plt.ylabel('Total Bill')
'''sns.lineplot(data=tips['day'],label='Days')
sns.lineplot(data=tips['total_bill'],label='Total Bill')'''
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(6,4))
sns.scatterplot(y=tips['day'],x=tips['tip'],size=tips['sex'])
plt.title('Days Vs Tips')
```

```python
#adding hue instead of size from above
plt.figure(figsize=(6,4))
sns.scatterplot(y=tips['day'],x=tips['tip'],hue=tips['sex'])
plt.title('Days Vs Tips')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
bill_day=tips[['day','time','total_bill']]
bill_day.groupby(['day','time']).sum()
```

```python
sns.boxplot(x='day',y='total_bill',data=bill_day,hue='time')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
#Tip Based on Dinner
tipdinner=tips[tips['time']=='Dinner'][['tip','time']]
```

```python
tiplunch=tips[tips['time']=='Lunch'][['tip','time']]
```

```python
tiplunch.head()
```

```python
plt.figure(figsize=(8,6))
#sns.distplot(a=tipdinner['tip'])
#sns.distplot(a=tiplunch['tip'])
#plt.subplots(1,2)
plt.subplot(1,2,1)
plt.hist(tipdinner['tip'])
plt.title('Tip Amount for Dinner Histogram')
plt.subplot(1,2,2)
plt.title('Tip Amount for Lunch Histogram')
plt.hist(tiplunch['tip'])

plt.show()
```

```python
#another way to do it

plt.figure(figsize=(8,6))
#sns.distplot(a=tipdinner['tip'])
#sns.distplot(a=tiplunch['tip'])
#plt.subplots(1,2)
plt.subplot(1,2,1)
sns.distplot(tipdinner['tip'],color='b')
plt.title('Tip Amount for Dinner Histogram')
plt.subplot(1,2,2)
plt.title('Tip Amount for Lunch Histogram')
sns.distplot(tiplunch['tip'],color='r')

plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
male=tips[tips['sex'] =='Male'][['total_bill','tip','smoker']]
female=tips[tips['sex'] =='Female'][['total_bill','tip','smoker']]
```

```python
male.tail()
```

```python
male['total_bill']
```

```python
female.tail()
```

```python
female['tip']
```

```python
plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
sns.scatterplot(x=male['total_bill'],y=male['tip'],hue=male['smoker'])
plt.title('Male- Total Bill vs Tip')
plt.subplot(1,2,2)
sns.scatterplot(x=female['total_bill'],y=female['tip'],hue=female['smoker'])
plt.title('Female- Total Bill vs Tip')
plt.show()
```

```python
sns.scatterplot(x=female['total_bill'],y=female['tip'],hue=female['smoker'])
plt.title('Female- Total Bill vs Tip')
plt.show()
```
