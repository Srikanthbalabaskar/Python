### Step 15. Calculate the mean windspeed for each month in the dataset.  
#### Treat January 1961 and January 1962 as *different* months.
#### (hint: first find a  way to create an identifier unique for each month.)

```python
datanew1=datanew.copy()
datanew1['Month_str']=datanew1['Month'].apply(lambda x: str(x))
datanew1['Year_str']=datanew1['Year'].apply(lambda x: str(x))
datanew1['Year-Month']=datanew1['Year_str']+'-'+datanew1['Month_str']
datanew1.head()
```

```python
datanew1.groupby('Year-Month').mean()
```
