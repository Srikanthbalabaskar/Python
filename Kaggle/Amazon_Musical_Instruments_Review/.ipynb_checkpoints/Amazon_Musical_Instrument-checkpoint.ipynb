# Description of Data Set

This file has reviewer ID , User ID, Reviewer Name, Reviewer text, helpful, Summary(obtained from Reviewer text),Overall Rating on a scale 5, Review time
Description of columns in the file:

reviewerID - ID of the reviewer, e.g. A2SUAM1J3GNN3B
asin - ID of the product, e.g. 0000013714
reviewerName - name of the reviewer
helpful - helpfulness rating of the review, e.g. 2/3
reviewText - text of the review
overall - rating of the product
summary - summary of the review
unixReviewTime - time of the review (unix time)
reviewTime - time of the review (raw)

```python
def first_helpful(helpful):
    return helpful.apply(lambda x:x.split(',')[0].split('[')[1])
def second_helpful(helpful):
    return helpful.apply(lambda x:x.split(',')[1].split(']')[0])
```

```python
music['foundHelpful']=first_helpful(music['helpful'])
```

```python
music['totalHelpful']=second_helpful(music['helpful'])
```

```python
def percent_helpful(a,b):
    y=float((a/b)*100)
    return round(y,2)
```

```python
music['foundHelpful']=music['foundHelpful'].astype('object')
music['totalHelpful']=music['totalHelpful'].astype('object')
```

```python
music.loc[music['overall']==5.0]
```

```python
def look_up(x):
    if x==5.0:
        return 'Equal to 5 Star'
    if x >=4.0 and x <5.0:   
        return 'Between 4 & 5 Star'
    if x >=3.0 and x <4.0:
        return 'Between 3 & 4 Star'
    if x >=2.0 and x <3.0:
        return 'Between 2 & 3 Star'
    if x<2.0:
        return 'less than 2 Star'
```

```python
music['ratingRange']= music['overall'].apply(look_up)
```

```python
musicReviewerName=music[['reviewerName','reviewText']].groupby('reviewerName').count().sort_values('reviewText',ascending=False).head(10)
musicReviewerName
```
