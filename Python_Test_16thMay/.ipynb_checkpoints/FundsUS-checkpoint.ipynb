### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv). 

### Step 3. Assign it to a variable called 

```python
called=pd.read_csv('https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv')
```

```python
called.head()
```

```python
called.shape
```

### Step 4.  What is the frequency of the dataset?

```python
called.groupby('Date').size()
```

### Step 5. Set the column Date as the index.

```python
called.set_index('Date',inplace=True)
```

```python
called.head()
```

### Step 6. What is the type of the index?

```python
called.index.dtype
#It is Object
```

### Step 7. Set the index to a DatetimeIndex type

```python
called.index=pd.to_datetime(called.index)
```

```python
called.index.dtype
```

### Step 8.  Change the frequency to monthly, sum the values and assign it to monthly.

```python
called['Datedup']=called.index
```

```python
called.head()
```

```python
called['Month']=called['Datedup'].dt.month_name()
```

```python
called.head()
```

```python
called.groupby(['Month']).sum()
```

### Step 9. You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.

### Step 10. Good, now we have the monthly data. Now change the frequency to year.

```python
called['Year']=called['Datedup'].dt.year
```

```python
called.groupby(['Year']).sum()
```
