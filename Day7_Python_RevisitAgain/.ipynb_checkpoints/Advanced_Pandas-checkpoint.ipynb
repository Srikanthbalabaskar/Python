# Apply 

Signature: df.apply(func, axis=0, broadcast=None, raw=False, reduce=None, result_type=None, args=(), **kwds) Docstring: Apply a function along an axis of the DataFrame.

Objects passed to the function are Series objects whose index is either the DataFrame's index (axis=0) or the DataFrame's columns (axis=1). By default (result_type=None), the final return type is inferred from the return type of the applied function. Otherwise, it depends on the result_type argument.

```python
#removing Nan(not a number)
df=df.dropna()
df
```

```python
#Before going into Apply function lets see how to Square the meal price

lst=[]
for i in df['Meal Price']:
    lst.append(i*i)
lst
```

```python
df['Meal Squared']=lst
```

```python
#Let us do the same thing with Apply function
def square(x):
    return x*x
df['Meal Squared Apply']=df['Meal Price'].apply(square)
df.tail()
```

```python
#the same can be done using Lambda
df['Meal Squared Lambda']=df['Meal Price'].apply(lambda x:x*x)
df.head()
```

```python
df['Meal Squared'].apply(lambda x:type(x)).value_counts()
```

```python
#Way to access row values
#df['Meal Squared']['(critera to access rows)']
df['Meal Squared'][0]    #gives the 1st row value
```

```python
df.loc['70000','Meal Squared']="Not a String"
df
```

```python
#Getting to know whether a String instance is available over a Particular Column
df['Meal Squared'][df['Meal Squared'].apply(lambda x:isinstance(x,str))].index
```

```python
# this will throw an error as we are trying to change a String also to int
#df['Meal Squared'] = pd.to_numeric(df['Meal Squared'])
df['Meal Squared'] = pd.to_numeric(df['Meal Squared'],errors='coerce') #70000 converted into Na
df.tail()
```

```python
#Fill Na fills Na values with the values we are giving
df['Meal Squared']=df['Meal Squared'].fillna(df['Meal Squared'].mean())
df.tail()  # mean gets filled  the meal squared na values
```

```python
#Fill median for Na value of 'Meal Squared Apply'
df['Meal Squared Apply']=df['Meal Squared Apply'].fillna(df['Meal Squared Apply'].median())
df.tail()
```

```python
#Changing Meal Squared Value to int
df['Meal Squared'].astype(int)
```

```python
#fill Na with 0's
df['Meal Squared Lamba']=df['Meal Squared Lambda'].fillna(0)
df.tail()
```

```python
#replacing 0's with median, note that in a Data Set 0 in most cases would be a valid value so think before replacing it
df['Meal Squared Lambda']=df['Meal Squared Lambda'].replace(0,df['Meal Squared Lambda'].median())
df.tail()
```

# Adding rows or Combining dataframes¶

```python
new_row = {'Order Id':'3224frefdsfd3r24',
           'Date':'2016-04-22',
           'Meal Id':'In423fdsfsf'}
```

```python
df_test=df.copy() # deep copy, when we change any value in df_test it wont affect df
```

```python
df_test =df_test.append(new_row,ignore_index=True)
```

```python
df_test.tail()
```

```python
#Create a new Row
new_row = {'Order Id':['3224frefdsfd3r24','frdsfdsf34434234'],
           'Date':['2016-04-22','2017-06-24'],
           'Meal Id':['In423fdsfsf','sdgsdgdsg']}
```

```python
add=pd.DataFrame(new_row)
add
```

```python
df_new=pd.concat([df_test,add],ignore_index=True)
df_new
```

```python
#Index gets generated as 0,1 for the newly added values
df_new_withoutignore = pd.concat([df_test,add])
df_new_withoutignore
```

```python
#Use reset index
df_new_withoutignore.reset_index()
```
